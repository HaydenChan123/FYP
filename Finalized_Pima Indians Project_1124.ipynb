#### Oversampling
When use any sampling technique ( specifically synthetic) you divide your data first and then apply synthetic sampling on the training data only. After you train you use the testing set ( which contains only original samples) to evaluate. The risk if you use your strategy is to have the original sample in training ( testing) and the synthetic sample ( that was created based on this original sample) in the testing ( training) set.



```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_features, Y, test_size=0.2, stratify = Y, random_state= 98754)
#888,88,67,123

from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=123)
X_train, Y_train = ros.fit_resample(X_train, Y_train)
print(X_train.shape)
print(X_test.shape)
print(Y_train.size)
print(Y_test.size)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import lightgbm as lgb
from lightgbm import LGBMClassifier
from sklearn.model_selection import StratifiedKFold
from scipy.stats import randint 
from scipy.stats import uniform 
models = []


dec_param_grid = {
    'max_depth': np.arange(2,12, 1)
}
rf_param_grid = {
    'max_depth': np.arange(2,12, 1),
    "n_estimators": np.arange(10, 22, 1),
              
              "min_samples_split": np.arange(15, 20, 1),
              "min_samples_leaf": np.arange(5, 20,1),
              "max_leaf_nodes": [20, 40],
              
    
    
    
}
xgb_param_grid = {

              'max_depth': randint(1, 11),
                          'subsample': uniform(0.25, 0.75),
                         'colsample_bytree': uniform(0.25, 0.75)

}
lgb_param_grid ={
     'max_depth': randint(1, 11),
                         'subsample': uniform(0.25, 0.75),
                         'colsample_bytree': uniform(0.25, 0.75)
    
    
}

#xg 125,233,527
#13
models.append(("Decision Tree:", dec_param_grid, DecisionTreeClassifier()))
models.append(("Random Forest:", rf_param_grid, RandomForestClassifier(max_depth = None, min_samples_leaf=8, min_samples_split=2,
                                                                       n_estimators = 20, random_state = 69)))

models.append(("eXtreme Gradient Boost:", xgb_param_grid, XGBClassifier( min_samples_leaf=8, min_samples_split=2, silent=True,
                                                                        n_estimators = 20, random_state = 876,num_leaves = 10)))

models.append(("Light Gradient Boost:", lgb_param_grid, LGBMClassifier(objective = 'binary',metric = 'binary_logloss',random_state=7, silent=True, n_jobs=4,
                                                                       num_leaves = 10)))
#LGB 1,5,7
print('Models appended...')
```
